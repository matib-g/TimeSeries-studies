```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import acf,pacf
from statsmodels.graphics.tsaplots import plot_acf
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

## MA Model

```python
def MA(n,c,burnin,params,epsilon,initial_values):
    result = initial_values
    for i in range((len(initial_values)-1),n):
        v = c+ epsilon[i] + sum(params[j]*epsilon[i-j-1] for j in range(len(params)))
        result.append(v)
        #result.append(c+sum([params[j]*result[i-j] for j in range(len(params))])+epsilon[i])
    return np.array(result[burnin:])
```

#  Task 1

```python
n=5000
burnin = 200
c=20
phi = 0.8
noise = np.random.normal(size=(n+burnin))
ts = MA(n+burnin,c,burnin,[phi],noise,[c])

plt.plot(ts)
```

```python
n=5000
burnin = 200
c=20
phi = 0.8
noise = np.random.normal(size=(n+burnin))
ts = MA(n+burnin,c,burnin,[phi],noise,[c])

fig,axes = plt.subplots(1,3,figsize=(15,5))
axes[0].plot(ts)
axes[0].set_title(f'Time series phi={phi} c={c}')
axes[1].plot(acf(ts),marker='o')
axes[1].set_title('ACF')
axes[2].plot(pacf(ts),marker='o')
axes[2].set_title('PACF')
```

```python
n=5000
burnin = 200
c=0
phi = -1
p_phi = 0.8
noise = np.random.normal(size=(n+burnin))
ts = MA(n+burnin,c,burnin,[phi,p_phi],noise,[c,c])

fig,axes = plt.subplots(1,3,figsize=(15,5))
axes[0].plot(ts)
axes[0].set_title(f'Time series phi={phi},p_phi={p_phi} c={c}')
axes[1].plot(acf(ts),marker='o')
axes[1].set_title('ACF')
axes[2].plot(pacf(ts),marker='o')
axes[2].set_title('PACF')
```
