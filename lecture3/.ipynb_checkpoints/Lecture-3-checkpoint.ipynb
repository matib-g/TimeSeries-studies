```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import SimpleExpSmoothing, Holt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_absolute_percentage_error
```

# Holt's method

### Importing GDP

```python
df = pd.read_csv('GDP.csv', index_col = 'DATE', parse_dates = True)
df.head()
```

```python
df.index.freq = 'QS'
df.info()
```

```python
df.plot(figsize = (12, 6)).autoscale(axis = 'x', tight = True)
```

```python
len(df)
```

### Splitting data into train and test

```python
ncut = int(0.8 * len(df))
ncut
```

```python
train_data = df.iloc[:ncut]
test_data = df.iloc[ncut:]
```

```python
ax = train_data.plot(figsize = (12, 6))
test_data.plot(ax = ax)
plt.legend(['Train', 'Test'])
```

### Simple Exponential Smoothing

```python
fitSES = SimpleExpSmoothing(train_data).fit()
fcastSES = fitSES.forecast(len(test_data)).rename('SES predict')
```

### Holt's method

```python
fitHolt = Holt(train_data, exponential = False).fit()
fcastHolt = fitHolt.forecast(len(test_data)).rename("Holt's predict")
```

```python
fitHolt.params['smoothing_level']
```

```python
fitHolt.params['smoothing_trend']
```

```python
fitHolt.params['damping_trend']
```

```python
ax = train_data.plot(figsize = (12, 6))
test_data.plot(ax = ax)
fcastSES.plot(ax = ax)
fcastHolt.plot(ax = ax)
plt.legend(['Train', 'Test', 'Predicted SES', 'Predicted Holt'])
```

### Comparing results

```python
print("SES error:")
print(100 * mean_absolute_percentage_error(test_data, fcastSES), "%")
```

```python
print("Holt's error:")
print(100 * mean_absolute_percentage_error(test_data, fcastHolt), "%")
```

### Holt's 2 (exponential)

```python
fitHolt = Holt(train_data, exponential = True).fit()
fcastHolt = fitHolt.forecast(len(test_data)).rename("Holt's predict")
```

```python
print(fitHolt.params['smoothing_level'])
print(fitHolt.params['smoothing_trend'])
```

```python
print("Holt's error:")
print(100 * mean_absolute_percentage_error(test_data, fcastHolt), "%")
```

### Holt 3 (damped trend)

```python
fitHolt = Holt(train_data, damped_trend = True).fit()
fcastHolt = fitHolt.forecast(len(test_data)).rename("Holt's predict")
```

```python
print(fitHolt.params['smoothing_level'])
print(fitHolt.params['smoothing_trend'])
print(fitHolt.params['damping_trend'])
```

```python
print("Holt's error:")
print(100 * mean_absolute_percentage_error(test_data, fcastHolt), "%")
```

### Holt 4 (exponential and damped)

```python
fitHolt = Holt(train_data, exponential = True, damped_trend = True).fit()
fcastHolt = fitHolt.forecast(len(test_data)).rename("Holt's predict")
```

```python
print(fitHolt.params['smoothing_level'])
print(fitHolt.params['smoothing_trend'])
print(fitHolt.params['damping_trend'])
```

```python
print("Holt's error:")
print(100 * mean_absolute_percentage_error(test_data, fcastHolt), "%")
```

# Holt-Winter's additive method

### Importing airline data

```python
df = pd.read_csv('airline_passengers.csv', index_col = 'Month', parse_dates = True)
df.head()
```

```python
df.index.freq = 'MS'
df.info()
```

```python
df.plot(figsize = (12, 6))
```

```python
resultSeasonal = seasonal_decompose(df['Thousands of Passengers'], model = 'multiplicative')
```

### Trend

```python
resultSeasonal.trend.plot(figsize = (12, 6))
```

### Seasonal component

```python
resultSeasonal.seasonal.plot(figsize = (12, 6))
```

### Residual

```python
resultSeasonal.resid.plot(figsize = (12, 6))
```

### Split into train and test

```python
train_data = df.iloc[:109]
test_data = df.iloc[109:]
```

```python
fitHoltWinter = ExponentialSmoothing(train_data['Thousands of Passengers'], trend = 'mul', seasonal = 'mul', seasonal_periods = 12).fit()
```

```python
fcastHoltWinter = fitHoltWinter.forecast(len(test_data)).rename("Holt-Winter's predict")
```

### Comparison of prediction and real values

```python
train_data['Thousands of Passengers'].plot(figsize = (12, 6), legend = True, label = 'Train')
test_data['Thousands of Passengers'].plot(legend = True, label = 'Test')
fcastHoltWinter.plot(legend = True, label = "Holt-Winter's predict")
```

```python
train_data['Thousands of Passengers'].plot(figsize = (12, 6), legend = True, label = 'Train')
test_data['Thousands of Passengers'].plot(legend = True, label = 'Test')
fcastHoltWinter.plot(legend = True, label = "Holt-Winter's predict", xlim = ['1958-01-01', '1961-01-01'])
```

# Stationarity

### Import sample data

```python
df = pd.read_csv('samples.csv', index_col = 0, parse_dates = True)
df.head()
```

### Plotting column A

```python
df['a'].plot(figsize = (12, 6), ylim = [0, 100], title = 'Stationary time series').autoscale(axis = 'x', tight = True)
```

### Augmented Dickey Fuller Test for stationarity

```python
adfuller(df['a'])
```

### Testing column B

```python
df['b'].plot(figsize = (12, 6), ylim = [0, 100], title = 'Stationary time series').autoscale(axis = 'x', tight = True)
```

```python
adfuller(df['b'])
```
