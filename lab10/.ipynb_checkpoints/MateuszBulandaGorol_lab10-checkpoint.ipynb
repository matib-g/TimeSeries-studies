# Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import train_test_split
```

```python
trainDF=pd.read_csv('train_6BJx641.csv',index_col='ID',parse_dates=True)
#df.index.freq='MS'
trainDF
```

```python
trainDF.plot(figsize=(15,8))
```

```python
le = LabelEncoder()
trainDF[['var2']] = le.fit_transform(trainDF[['var2']])
trainDF
```

```python
#n = 0.8
#split = int(n*len(trainDF))
#split
```

```python
y = trainDF.electricity_consumption
y
```

```python
features = ['temperature','var1','pressure','windspeed','var2']
```

```python
X = trainDF[features]
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0, train_size=0.8)
```

## Random Forest Regressor

```python
rfr_score = {'n_estimators':0, 'max_depth':0, 'train_score':0, 'test_score':0}
for n in np.arange(100, 1000, 100):
    print(f'number of trees: {n}')
    for i in range(1,10):
        forest_model = RandomForestRegressor(n_estimators=n, max_depth=i)
        forest_model.fit(train_X, train_y)
        train_score = forest_model.score(train_X, train_y)
        test_score = forest_model.score(val_X, val_y)
        print(f'max depth: {i}, test_score: {test_score}')
        if(test_score > rfr_score['test_score']):
            rfr_score['n_estimators'] = n
            rfr_score['max_depth'] = i
            rfr_score['train_score'] = train_score
            rfr_score['test_score'] = test_score
```

```python
rfr_score
```

```python
forest_model = RandomForestRegressor(n_estimators=rfr_score['n_estimators'], max_depth=rfr_score['max_depth'])
forest_model.fit(train_X, train_y)
rfr_preds = forest_model.predict(val_X)
print(mean_absolute_percentage_error(val_y, rfr_preds))
```

## Gradient Boosting Regressor

```python
gbr_score = {'n_estimators':0, 'max_depth':0, 'train_score':0, 'test_score':0}
for n in np.arange(100, 1000, 100):
    print(f'number of trees: {n}')
    for i in range(1,10):
        gbr_model = GradientBoostingRegressor(n_estimators=n, max_depth=i)
        gbr_model.fit(train_X, train_y)
        train_score = gbr_model.score(train_X, train_y)
        test_score = gbr_model.score(val_X, val_y)
        print(f'max depth: {i}, test_score: {test_score}')
        if(test_score > gbr_score['test_score']):
            gbr_score['n_estimators'] = n
            gbr_score['max_depth'] = i
            gbr_score['train_score'] = train_score
            gbr_score['test_score'] = test_score
```

```python
gbr_score
```

```python
gbr_model = RandomForestRegressor(n_estimators=gbr_score['n_estimators'], max_depth=gbr_score['max_depth'])
gbr_model.fit(train_X, train_y)
gbr_preds = gbr_model.predict(val_X)
print(mean_absolute_percentage_error(val_y, gbr_preds))
```
