# SARIMA model

```python
df = pd.read_csv('vacation.csv', index_col='Month', parse_dates=True)
df.index.freq='MS'
```

```python
sd_results = seasonal_decompose(df)
sd_results.plot()
```

```python
n = 12
train = df[:-n]
test = df[-n:]
start = len(train)
stop = len(df)-1
```

```python
auto_arima(train,seasonal=True,m=12,trace=True).summary()
```

```python
SarimaxModel = SARIMAX(train, order=(1, 1, 1), seasonal_order=(1,0,1,12)).fit()
predSarimax = SarimaxModel.predict(start=start, end=stop, dynamic=False).rename('SARIMA')
test.plot(figsize=(12,6), legend=True)
predSarimax.plot(legend=True)
```

```python
errors['Sarima'] = mean_absolute_percentage_error(test,predSarimax)
errors['Sarima']
```

## Predictions 12 months in the future using SARIMAX model

```python
futureSarimaxModel = SARIMAX(df, order=(1, 1, 1), seasonal_order=(1,0,1,12)).fit()
futureSarimax = futureSarimaxModel.predict(start=stop, end=stop+11, dynamic=False).rename('SARIMA')
futureSarimax.index= pd.to_datetime(futureSarimax.index)
test.plot(figsize=(12,6), legend=True)
futureSarimax.plot(legend=True)
```

# Holt's method

## Let's check which pair of parameter get best results.

```python
params = ['add', 'mul']
HoltErrors = {}
for trend in params:
    for seasonal in params:
        ExpModel = ExponentialSmoothing(train, trend=trend, seasonal=seasonal).fit()
        predHolt = ExpModel.forecast(n).rename(f'Holts-Winter ({trend}, {seasonal})')
        HoltErrors[f'Holts-Winter ({trend}, {seasonal})'] = mean_absolute_error(test,predHolt)
```

```python
bestHolt = min(HoltErrors, key=HoltErrors.get)
bestHolt
```

```python
ExpModel = ExponentialSmoothing(train, trend='mul', seasonal='add').fit()
predHolt = ExpModel.forecast(n).rename('Holts-Winter mul add')
errors['Holts-Winter (mul, add)'] = mean_absolute_percentage_error(test,predHolt)
ax = test.plot(legend=True, figsize=(12,6))
predHolt.plot(ax=ax,legend=True)
```
